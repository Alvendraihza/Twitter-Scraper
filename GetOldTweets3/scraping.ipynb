# The Function

```python
import GetOldTweets3 as got
import pandas as pd

def get_tweets(kata_kunci,
               tanggal_mulai = "2020-01-01", 
               tanggal_akhir = "2020-05-31", 
               di_daerah_rad = "Jakarta",
               batas_mencari = 100):
    
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch(kata_kunci)\
                                               .setSince(tanggal_mulai)\
                                               .setUntil(tanggal_akhir)\
                                               .setNear(di_daerah_rad)\
                                               .setMaxTweets(batas_mencari)
    
    tweet = got.manager.TweetManager.getTweets(tweetCriteria)
    
    text_tweets = [[tw.username,
                    tw.text,
                    tw.date,
                    tw.retweets,
                    tw.favorites,
                    tw.mentions,
                    tw.hashtags] for tw in tweet]
    
    df_state = pd.DataFrame(text_tweets, columns = ['User', 
                                                    'Text', 
                                                    'Date', 
                                                    'Favorites', 
                                                    'Retweets', 
                                                    'Mentions',
                                                    'Hashtags'])
    
    return df_state
```

# Result

```python
kata_kunci = ["corona", "#corona", 
              "corona virus", "virus corona", "#coronavirus", "#viruscorona",
              "covid-19", "covid19", "#covid_19", "#covid19",
              "psbb", "#psbb",
              "di rumah aja", "#dirumahaja",
              "wfh", "#wfh",
              "work from home", "#workfromhome",
              "belajar di rumah", "#belajardirumah"]

hasil = pd.DataFrame()
hasil = get_tweets(kata_kunci[0])
   
for i in kata_kunci[1:]: 
    df = get_tweets(i)
    hasil = pd.concat([hasil, df], axis = 0)

hasil.drop_duplicates(keep = False, inplace = True) 
hasil.to_csv('+ indo corona.csv')
hasil
```
